```python
import pandas as pd

# Load full dataset
df = pd.read_csv("dataset/train.csv")

# Split into 3 equal parts
n = len(df)
batch_size = n // 3

for i in range(3):
    start = i * batch_size
    end = (i + 1) * batch_size if i < 2 else n
    df_batch = df.iloc[start:end]
    df_batch.to_csv(f"dataset/train_batch_{i+1}.csv", index=False)
    print(f"Created dataset/train_batch_{i+1}.csv with {len(df_batch)} rows")
```
